```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cost_function(x) :
    return (x+3)**2
```

```python
def gradient(x) :
    return 2*(x+3)
```

```python
x_values=[]
y_values=[]
num_iterations=100
```

```python
x_initial=2
x=x_initial
learning_rate=0.1
tolerance=1e-6
```

```python
for i in range(num_iterations):
    x_values.append(x)
    y_values.append(cost_function(x))
    x=x-learning_rate*gradient(x)
    print(f'Iteration {i+1} : x value: {x} Cost Function:{cost_function(x)} ')
   
    
```

```python
print(f'Optimal x : {x}')
```

```python
x_coordinates = np.linspace(-5,5,100)
plt.plot(x_coordinates,cost_function(x_coordinates))
plt.plot(x_values,y_values,'ro')
plt.xlabel("Iteration")
plt.ylabel("Cost Function Value")
plt.title("Gradient Descent Trace of Cost Function")
plt.show()
```
